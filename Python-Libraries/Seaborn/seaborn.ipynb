```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("Libraries Imported")
```

```python
df =pd.read_csv('Used_Bikes.csv')
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
brand_name = list(df['brand'].value_counts().keys())
bike_count = list(df['brand'].value_counts().values)
```

```python
plt.bar(brand_name,bike_count)
plt.xlabel("Brands")
plt.ylabel("Bikes_Counts")
plt.show()
```

```python
sns.barplot(x=brand_name,y=bike_count)
plt.xlabel("Brands")
plt.ylabel("Bikes_Counts")
plt.show()
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
sns.relplot(x ='tip',y = 'total_bill',data =df)
plt.show()


```

```python
sns.relplot(x ='tip',y = 'total_bill',data =df,hue ='sex')
plt.show()


```

```python
sns.relplot(x ='tip',y = 'total_bill',data =df,hue ='sex',style = 'smoker')
plt.show()


```

```python
sns.relplot(x ='tip',y = 'total_bill',data =df,hue ='sex',style = 'smoker',col ='day')
plt.show()


```

```python
sns.relplot(x ='tip',y = 'total_bill',data =df,hue ='time',col ='day')
plt.show()


```

```python
sns.relplot(x ='tip',y = 'total_bill',data =df,hue ='time',col ='day',kind ='line')
plt.show()


```

```python
sns.relplot(x ='tip',y = 'total_bill',data =df,col ='day',kind ='line')
plt.show()
```

```python
sns.relplot(x ='size',y = 'total_bill',data =df,col ='day',kind ='line')
plt.show()
```

Catgorical PLot
"boxen", "point", "bar", or "count"

```python
sns.catplot(x='smoker',y='total_bill',data =df,hue ='sex')
plt.show()
#STRIP CHART
```

```python
sns.catplot(x='smoker',y='total_bill',data =df,hue ='sex',kind ='swarm')
plt.show()
# SWARN CHART
```

```python
sns.catplot(x='smoker',y='total_bill',data =df,hue ='sex',kind ='box')
plt.show()
#BOX CHART
```

```python
sns.catplot(x='smoker',y='total_bill',data =df,hue ='sex',kind ='point')
plt.show()
#Point chart
```

```python
sns.catplot(x='smoker',y='total_bill',data =df,hue ='sex',kind ='boxen')
plt.show()
#Boxen Chart
```

```python
sns.catplot(x='smoker',y='total_bill',data =df,hue ='sex',kind ='violin')
plt.show()
#Violin Chart
```
