```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']

plt.pie(country_population,labels=country_name,autopct="%2.2f%%")
plt.show()
```

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
explosion = [0,0,0.3,0,0]
plt.pie(country_population,labels=country_name,autopct="%1.2f",explode=explosion)
plt.show()
```

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
explosion = [0,0,0.3,0,0]
plt.pie(country_population,labels=country_name,autopct="%1.2f",explode=explosion,wedgeprops={'ec':'black'})
plt.show()
```

```python

plt.figure(figsize=(6,12))
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']
explosion = [0,0,0.3,0,0]
plt.pie(country_population,labels=country_name,autopct="%1.2f",explode=explosion,wedgeprops={'ec':'black','lw':3},shadow=True)
plt.savefig("country_pipplot")
plt.show()
```
